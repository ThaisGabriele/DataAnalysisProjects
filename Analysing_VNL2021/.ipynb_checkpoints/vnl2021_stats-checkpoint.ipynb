## 2. Exploratory Data Analysis (EDA)

### Dataset
* All the following files contains the data collected of the players perfomance in all the competition according to the following skills (attacking, blocking, digging, receiving, scoring, serving, setting). All these data was extracted from the [official website](https://en.volleyballworld.com/volleyball/competitions/vnl-2021/statistics/) of the competition using simple web scrapping techniques that consists in reading the tables as dataframes.


```python
# reading files
players = pd.read_csv("datasets/players.csv", index_col=0)
scorers = pd.read_csv("datasets/best_scorers.csv",index_col=0)
attackers = pd.read_csv("datasets/best_attackers.csv", index_col=0)
blockers = pd.read_csv("datasets/best_blockers.csv", index_col=0)
servers = pd.read_csv("datasets/best_servers.csv", index_col=0)
setters = pd.read_csv("datasets/best_setters.csv", index_col=0)
diggers = pd.read_csv("datasets/best_diggers.csv", index_col=0)
receivers = pd.read_csv("datasets/best_receivers.csv", index_col=0)
```

## 4. Data Visualization 

* Guided by the following questions in this section we will try to make a better visualization of the data and get some insights to possibily answer this questions

    * Questions 
        - Which team scored better in each skills? 
        - Which team had the best attacking efficiency?
        - Which team had the best passing?
        - Which team had the best blocking performance?
        - Can we find out which team that possibly won the competition? 

### 4.1 General scoring by Teams

```python
# attack points
att_points = (scorers.filter(items=['Team', 'AttackPoints'])
                     .groupby('Team')
                     .sum()
                     .sort_values('AttackPoints',ascending=False)
                     .head(10)
             )
```

```python
att_points.head()
```

```python
# block points
blk_points = (scorers.filter(items=['Team', 'BlockPoints'])
                     .groupby('Team')
                     .sum()
                     .sort_values('BlockPoints',ascending=False)
                     .head(10)
             )
```

```python
blk_points.head()
```

```python
# serve points
serve_points = (scorers.filter(items=['Team', 'ServePoints'])
                       .groupby('Team')
                       .sum()
                       .sort_values('ServePoints',ascending=False)
                       .head(10)
               )
```

```python
serve_points.head()
```

```python
fig, axes = plt.subplots(1,3, figsize=(13,5))
fig.suptitle("Top 10 Scoring Points by Skills")

sns.barplot(ax=axes[0], y=att_points.index, x=att_points['AttackPoints'])

sns.barplot(ax=axes[1],y=blk_points.index, x=blk_points['BlockPoints'])

sns.barplot(ax=axes[2],y=serve_points.index, x=serve_points['ServePoints'])

plt.show()
```

- *Analysing the above plots we can see that the turkish team (TUR) and usa team (USA) appears in all of the top 5 scoring skills (attacking, blocking and serving). Brazil (BRA) appears in only two.* 
- *In attacking, Brazil e Turkey have the two best numbers of attack points*
- *In blocking, the brazilian team*
- *In serving, the USA team*

### 4.2 Skill: attack
* In this subsection, we firstly try to verify the columns that are related to attacks efficiency percentage. The we check the attack efficiency by team considering the 'success%' column


**Checking the numbers of total attack attempts**

```python
total_att = (attackers.groupby(['Team'])
                    .sum()
                    .sort_values('TotalAttempts',ascending=False))
```

```python
total_att
```

```python
plt.figure(figsize=(14,8))

sns.barplot(x = 'TotalAttempts', 
            y = total_att.index, 
            data = total_att, 
            label = 'Total Attempts')

plt.show()
```

*-The japanese team (JPN) has the highest number of attack attempts*

**Checking attacking efficiency by 'success%' column (Success% = AttackPoints / TotalAttempts)**

```python
# dataframe with only two columns: team and success
attack_success = attackers.filter(items=['Team','Success%'])
```

```python
attack_success.head()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=attack_success, x='Team', y='Success%')

sns.despine(left = True, bottom = True)
plt.show()
```

*We can see that some of the boxplots median stays near to the 40%*

```python
attack_median = attack_success.groupby('Team').median()
```

```python
(attack_median.loc[attack_median['Success%'] >= 40.0]
              .sort_values('Success%',ascending=False)
)
```

*- CHN, BRA, TUR and USA had the best attacking efficiency considering the median of their attacking success*

### 4.3 Skill: Passing
* In this subsection, we firstly try to verify the columns that are related to receiving efficiency percentage. Then we check the receiving efficiency by team considering the 'success%' column. Next, we make all the necessary adjusts of the data to make a great chart.


```python
receivers.head()
```

```python
receivers.loc[receivers['Team'] == 'BRA']
```

```python
rec_success = receivers.filter(items=['Team','Success%','TotalAttempts'])
```

```python
rec_success
```

```python
# selecting only the rows with more than 5 receiving attempts 
rec_success = rec_success.loc[rec_success['TotalAttempts'] > 5.0]
```

*Considering that not every player in the court usually makes the passing. We filter the rows by the numbers of  total attempts of passing made in the competition*

```python
rec_success
```

```python
plt.figure(figsize=(16,4))

sns.boxplot(data=rec_success, 
            x='Team', 
            y='Success%')

plt.show()
```

```python
receivers_med = rec_success.groupby('Team').median()
```

```python
receivers_med
```

```python
receivers_med.median()
```

```python
receivers_med = (receivers_med.loc[receivers_med['Success%'] >= 30.0]
                              .sort_values('Success%', ascending=False)
                )
```

```python
plt.figure(figsize=(10,5))

sns.barplot(data=receivers_med,
           x=receivers_med.index,
           y='Success%')
plt.show()
```

- *Looking at the bloxplot distributions and the bar plot. The japanese team shows the best efficiency in passing.*
- *When comparing the other team like USA and TUR the difference is very low*

## Skill: block
* In this subsection, by analysing the data we decided check the block performance by comparing the number of kill blocks/block points made.
* So, we considered that the highest number of kill blocks reflects in a good block performance 

```python
blockers.head()
```

```python
blks = blockers.filter(items=['Team','Blocks'])
```

```python
blocks = (blks.groupby("Team")
              .sum()
              .sort_values('Blocks',ascending=False)
         )
```

```python
blocks
```

```python
plt.figure(figsize=(10,5))

sns.barplot(data=blocks,
           x=blocks.index,
           y='Blocks')

plt.show()
```

```python
blocking = blks.loc[blks['Blocks'] > 0 ]
```

```python
plt.figure(figsize=(16,4))

sns.boxplot(data=blocking, 
            x='Team', 
            y='Blocks')


plt.show()
```

- *Looking at the bloxplot distributions and the bar plot. The brazilian team shows the best blocking performance*
